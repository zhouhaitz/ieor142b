# Hannah Scratch CNN Manual Poster Test

This notebook runs Hannah's trained scratch CNN checkpoint on one movie poster.

Default behavior:
- load `hannah_scratch_cnn.pt` from this folder
- choose one deterministic random poster from `cleaned/downloaded_posters`
- print predicted labels above threshold and top genre probabilities

If the checkpoint is missing, run the full-training cell in `experimentation/hannah.ipynb` first.

```python
from pathlib import Path
import subprocess
import sys


# Set this manually if your notebook kernel starts outside the repo.
ROOT_OVERRIDE = Path('/Users/user/ieor142b')


def find_repo_root(start: Path | None = None) -> Path:
    start = (start or Path.cwd()).resolve()
    candidates = [
        ROOT_OVERRIDE,
        start,
        *start.parents,
        Path('/content/ieor142b_runtime'),
        Path('/content/drive/MyDrive/ieor142b'),
    ]
    seen = set()
    for candidate in candidates:
        candidate = candidate.expanduser().resolve()
        if str(candidate) in seen:
            continue
        seen.add(str(candidate))
        if (candidate / 'cleaned' / 'MovieGenre_clean_with_images_full.csv').is_file():
            return candidate
    raise RuntimeError('Could not find the ieor142b repo root.')


ROOT = find_repo_root()
TEST_SCRIPT = ROOT / 'experimentation' / 'hannah' / 'checkpoints' / 'hannah_test.py'
CHECKPOINT = ROOT / 'experimentation' / 'hannah' / 'checkpoints' / 'hannah_scratch_cnn.pt'

print('Repo root:', ROOT)
print('Test script:', TEST_SCRIPT)
print('Expected checkpoint:', CHECKPOINT)
```

```python
# Edit these if you want a specific poster or threshold.
IMAGE_PATH = None  # Example: ROOT / 'cleaned' / 'downloaded_posters' / '10040.jpg'
THRESHOLD = None   # Example: 0.40
TOP_K = 8
SEED = 142
```

```python
cmd = [sys.executable, str(TEST_SCRIPT), '--top-k', str(TOP_K), '--seed', str(SEED)]
if IMAGE_PATH is not None:
    cmd.extend(['--image', str(IMAGE_PATH)])
if THRESHOLD is not None:
    cmd.extend(['--threshold', str(THRESHOLD)])

print('Running:', ' '.join(cmd))
subprocess.run(cmd, cwd=ROOT, check=True)
```
